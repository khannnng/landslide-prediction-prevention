```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# all columns, limit 200 rows
pd.set_option("display.max_columns", None) 
pd.set_option("display.max_rows", 200)

# round up DISPLAY 4 decimal
pd.options.display.float_format = "{:,.4f}".format

import time

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, StratifiedKFold, cross_validate, cross_val_score, cross_val_predict
from sklearn.model_selection import GridSearchCV
from sklearn import metrics

from sklearn.metrics import ( 
    f1_score, 
    accuracy_score, 
    recall_score, 
    precision_score, 
    confusion_matrix, 
    roc_auc_score, 
    roc_curve,
    plot_confusion_matrix, 
    classification_report, 
    precision_recall_curve,
    make_scorer,
 
) 

```

# 1. Load data

```python
path = 'df/'
df = pd.read_csv(path+'Train.csv')
df.drop(columns='Sample_ID', inplace=True)
```

```python
X = df.drop(columns='Label')
Y = df['Label']
X.head()
```

```python
# add preceding 0 to single digit column
cellid    = [name.split('_')[0] for name in X.columns]
cellid = ['0' + id_num if (len(id_num) == 1) else id_num for id_num in cellid]
attribute = [name.split('_')[1] for name in X.columns]

X.columns = ['_'.join(name) for name in zip(cellid,attribute)]
```

```python
# Group each pixel's "channels" sequently
X = X[sorted(X.columns)]
X.head()
```

### Min Max scaler

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, stratify = Y, test_size = 0.1, random_state=14)
```

```python
# normal scale
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### Reshape

```python
x_train = x_train.reshape(len(x_train), 5, 5, 9)
x_train.shape
```

```python
x_test = x_test.reshape(len(x_test), 5, 5, 9)
x_test.shape
```

# Build model

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
```

```python
# specify input dimensions of each image
img_rows, img_cols = 5, 5
depth = 9
input_shape = (img_rows, img_cols, depth)

# batch size, number of classes, epochs
batch_size = 128
num_classes = 2
epochs = 10
```

### Model 1

```python
# model
model = Sequential()

# a keras convolutional layer is called Conv2D
# help(Conv2D)
# note that the first layer needs to be told the input shape explicitly

# first conv layer
model.add(Conv2D(64, kernel_size=(3, 3),
                 activation='relu',
                 padding='same',
                 input_shape=input_shape))
# model.add(Dropout(0.5))

# second conv layer
model.add(Conv2D(64, kernel_size=(3, 3), 
                 activation='relu',
                padding='same'))

model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# flatten and put a fully connected layer
model.add(Flatten())
model.add(Dense(128, activation='relu')) # fully connected
model.add(Dropout(0.1))

# softmax layer
model.add(Dense(1, activation='sigmoid'))

# model summary
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
# make prediction
threshold = 0.5
yhat_train = model.predict(x_train) > threshold
yhat_test = model.predict(x_test) > threshold
```

```python
# baseline CNN model performance
fig, [ax1,ax2] = plt.subplots(ncols=2, figsize=(8,4))
sns.heatmap(confusion_matrix(y_train, yhat_train), annot=True, fmt='.0f', ax=ax1, cbar=False)
sns.heatmap(confusion_matrix(y_test, yhat_test), annot=True, fmt='.0f', ax=ax2, cbar=False)
print(classification_report(y_train, yhat_train))
print(classification_report(y_test, yhat_test))
```

```python
# function to plot precision/recall curve
def pre_rec_curve(y_train, y_scores, threshold=0.5):

    precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
    thresholds = np.insert(thresholds, 0, 0)
    F1_score = 2*precisions*recalls/(precisions+recalls)


    # Plot values of precisions, recalls, and thresholds
    plt.figure(figsize=(10,7))
    plt.plot(thresholds, precisions, 'b--', label='precision')
    plt.plot(thresholds, recalls, 'g--', label = 'recall')
    plt.plot(thresholds, F1_score, 'r--', label = 'f1 score')

    plt.xlabel('Threshold')
    plt.legend(loc='upper left')
    plt.ylim([0.4,1])
    plt.axvline(threshold)
    plt.show()
```

```python
yhat_train = model.predict(x_train)
pre_rec_curve(y_train, yhat_train, threshold=0.32)
```

```python
# make prediction with optimal threshold
threshold = 0.3
yhat_train = model.predict(x_train) > threshold
yhat_test = model.predict(x_test) > threshold
```

```python
# baseline CNN model performance
fig, [ax1,ax2] = plt.subplots(ncols=2, figsize=(8,4))
sns.heatmap(confusion_matrix(y_train, yhat_train), annot=True, fmt='.0f', ax=ax1, cbar=False)
sns.heatmap(confusion_matrix(y_test, yhat_test), annot=True, fmt='.0f', ax=ax2, cbar=False)
print(classification_report(y_train, yhat_train))
print(classification_report(y_test, yhat_test))
```
